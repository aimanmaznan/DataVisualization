```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\user\Desktop\PortfolioProject2023\Data Visualization\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
#choosing theme
print(plt.style.available)
plt.style.use('classic')
```

```python
#Line Plot
df.plot(kind = 'line',title = 'Ice Cream Ratings',xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
#seperate Line Plot
df.plot(kind = 'line', subplots = True)
```

```python
#Bar Plot
df.plot(kind = 'bar')
```

```python
#Specific Column Bar Plot
df['Flavor Rating'].plot(kind = 'bar')
```

```python
#Stacked Bar Plot
df.plot(kind= 'bar',stacked = True)
```

```python
#Horizontal Bar Plot
df.plot.barh()
```

```python
#Stacked Horizontal Bar Plot
df.plot.barh(stacked = True)
```

```python
#Scatter Plot
df.plot.scatter(x = 'Texture Rating',y = 'Overall Rating', s = 500, c = 'Yellow')
```

```python
#Histogram
df.plot.hist(bins =10)
```

```python
#Boxplot
df.boxplot()
```

```python
#Area Plot
df.plot.area(figsize = (10,5))
```

```python
#Pie Chart
df.plot.pie(y = 'Flavor Rating', figsize = (10,10))
```
